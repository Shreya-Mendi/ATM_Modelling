```python
import pandas as pd
import zipfile

zip_path = "atm Data-selected.zip"
dfs = {}

with zipfile.ZipFile(zip_path, "r") as z:
    for filename in z.namelist():
        if filename.endswith(".parquet"):
            with z.open(filename) as f:
                dfs[filename] = pd.read_parquet(f)

# Access individual dataframes
print(dfs.keys())
for name, df_part in dfs.items():
    print(f"\n=== {name} ===")
    print("Shape:", df_part.shape)
    print("Columns:", list(df_part.columns))
    print(df_part.head(3))  # show first 3 rows

```
